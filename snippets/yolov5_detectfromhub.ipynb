# Default Example

```python
import torch
print(torch.__version__)
```

```python
# Set the custom directory for PyTorch Hub
custom_dir = 'E:/Code/Unviveral_SFR_Target_Finder'  # Specify your custom path here
torch.hub.set_dir(custom_dir)

# Now, when you load a model via PyTorch Hub, it will use the custom directory
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
from PIL import Image
import requests

# Load an image
img_url = 'https://th-thumbnailer.cdn-si-edu.com/ii_ZQzqzZgBKT6z9DVNhfPhZe5g=/fit-in/1600x0/filters:focal(1061x707:1062x708)/https://tf-cmsv2-smithsonianmag-media.s3.amazonaws.com/filer_public/55/95/55958815-3a8a-4032-ac7a-ff8c8ec8898a/gettyimages-1067956982.jpg'  # Replace with your image URL
img = Image.open(requests.get(img_url, stream=True).raw)

# Perform detection
results = model(img)

# Results show
results.show()
```

```python
# Print results
results.print()  

# Get bounding boxes for plotting
bboxes = results.xyxy[0]  # Bounding boxes in xyxy format
print(bboxes)
```
